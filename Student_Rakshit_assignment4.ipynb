```python
#Initialize
u='State-Place'
x='Year'
y='Population'
z='Rakshit - Rank'
rows_so_far=0
c=0
```

```python
#Initialize 2D list
pool=[]
pool.append([])
```

```python
#iterating through the csv file
for row in reader:
    if rows_so_far==0:
        rows_so_far+=1
        header=row
        
        for j in range(0,4):
            if j==0:
                pool.append([])
                pool[0].append(u)
            if j==1:
                pool[0].append(x)
            if j==2:
                pool[0].append(y)
            if j==3:
                pool[0].append(z)
    else:
        for i in range(len(row)-2):
            a=len(pool)
            if not row==[]:
                if i==0 or i>=1:
                    item=copy.deepcopy(row)
                    r=copy.deepcopy(row)
                    
                    #column values
                    
                    for j in range(0,4):
                        if item[i+2]!='':
                            if j==0:
                                
                                #state and city value
                                
                                r[0]=item[j+1]+'-'+item[j]
                                pool.append([])
                                pool[a-1].append(r[0])
                                
                                #year value
                                
                            if j==1:
                                pool[a-1].append(int(header[i+2]))
                                    
                                #population value
                                    
                            if j==2:
                                if item[i+2]=='':
                                    pool[a-1].append(int(0))
                                else:
                                    pool[a-1].append(int(item[i+2]))
                                
                                #initialize rank to zero
                                        
                            if j==3:
                                pool[a-1].append(int(0))
                                    
    rows_so_far+=1
                            
                
                                    
```

```python
#add header to the list
list1=[]
list1.append([])
list1[0]=pool[0]
list1[1:a-1]=list[0:a-2]
```

```python
mycsv=csv.writer(open(r'C:\\Users\\raksh\\OneDrive\\Desktop\\Semester 2\\Data Viz\\Assignments\\Assignment 4\\Rakshit.CSV','w',newline=''))
for row in list1:
    #fetching row index
    e=list1.index(row)
    #comparing the record values
    if row[1]!=c and e!=0:
        v=1
        c=row[1]
        row[3]=v
    else:
        if row[1] == c and e!=0:
            v+=1
            row[3]=v
    #write row to csv
    mycsv.writerow(row)
    
```
